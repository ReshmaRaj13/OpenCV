```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#matlplotlib inline
from IPython.display import Image
```

```python
# Read in an image
image = cv2.imread("home.jpg", 1)

# Display the original image
plt.imshow(image[:,:,::-1])
image.shape
```

## drawing a line

```python
imgline = image.copy()
cv2.line(imgline, (100, 100), (300, 100),(0,255,0), thickness = 10, lineType = cv2.LINE_AA)
plt.imshow(imgline[:,:,::-1])
```

# drawing a circle

```python
imgcircle= image.copy()
cv2.circle(imgcircle, (200, 200), 100,None, thickness = 4, lineType = cv2.LINE_AA)
plt.imshow(imgcircle, cmap = 'gray')
```

# drawing a rectangle

```python
imgrec= image.copy()
cv2.rectangle(imgrec, (100, 100), (200,350),(0,0,255), thickness = 4, lineType = cv2.LINE_8)
plt.imshow(imgrec[:,:,::-1])
```

# Adding a text

```python
imageText = image.copy()
text = "It's a home"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 255)
fontThickness = 2
cv2.putText(imageText, text, (100, 50), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);
plt.imshow(imageText[:, :, ::-1])           
```
